```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
file = "student-mat.csv"
data_sm = pd.read_csv(file)
print("read file complete")
nRow, nCol = data_sm.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
data_sm.describe()
```

```python
titres = data_sm.columns
print(titres[2])
print(data_sm.columns[29])
```

```python
print(titres)
```

```python
for i in titres :
    nb = data_sm[i].dropna()
    print(i," : ", nb.count())

```

```python
print("None NaN")
```

```python
ig,ax=plt.subplots()
ax.set_ylabel('Nbr personnes')
ax.set_title('Nbr de personnes par age')
ax.set_xlabel('Age')
data_sm[titres[2]].value_counts().plot.bar()
```

```python
ig,ax=plt.subplots()
ax.set_ylabel('Nbr personnes')
ax.set_title('Nbr d\'absence')
ax.set_xlabel('Nb absence')
nb_absence = data_sm[titres[29]].value_counts()
nb_absence.sort_index(inplace=True)
nb_absence.plot.bar()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=data_sm)
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=data_sm, y='absences', x='age')
```
