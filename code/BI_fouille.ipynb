```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
file = "student-mat.csv"
data_sm = pd.read_csv(file)
print("read file complete")
nRow, nCol = data_sm.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
ig,ax=plt.subplots()
ax.set_ylabel('Nbr personnes')
ax.set_title('Nbr de personnes par age')
ax.set_xlabel('Age')
trie = data_sm[titres[2]].value_counts()
trie.sort_index(inplace=True)
trie.plot.bar()
```
